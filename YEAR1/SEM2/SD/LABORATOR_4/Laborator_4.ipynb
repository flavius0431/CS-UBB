```python
#ex1
t=var('t')
k=var('k')
x0=var('x0')
x(t)=function('x')(t)
eqd1=diff(x,t)==-k*x
eqd2=x(0)==x0
syst=[eqd1,eqd2]
show(syst)

```

```python
ans(t,k,x0)=desolve(eqd1,[x,t],[0,x0])
```

```python
plot([ans(t,0.1,10),ans(t,0.2,10),ans(t,0.3,10),ans(t,0.4,10)],t,0,5)
```

```python
k=var('k')
ec=ans(t,k,x0)==x0/2
show(ec)
```

```python
anss=var('anss')
anss=solve(ec,k)
show(anss)
```

```python
k1(t)=anss[0].rhs()
numerical_approx(k1(5730))
```

```python
kC14=k1(5730)
ecc=ans(t,kC14,1)==0.2
sol=solve(ecc,t)
```
