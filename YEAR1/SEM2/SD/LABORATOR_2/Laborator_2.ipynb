```python
#2x^2y' = x^2 + y^2
reset()
x=var('x')
y=function('y')(x)
eqd=2*x^2*diff(y,x)==x^2+y^2
eqd
```

```python
show(eqd)
```

```python
desolve(eqd,y)
```

```python
sol = desolve(eqd,y)
show(sol)
```

```python
yy=var('yy')
sol.substitute(y(x)==yy)
```

```python
_C=var('_C')
f(x,yy,_C)=sol.substitute(y(x)==yy)
f(x,yy,_C)
```

```python
implicit_plot(f(x,yy,1),(x,-2,2),(yy,-2,2))
```

```python
g=implicit_plot(f(x,yy,1),(x,-2,2),(yy,-2,2))
for i in [-20..20]:
    g1=implicit_plot(f(x,yy,i/5),(x,-2,2),(yy,-2,2))
    g=g+g1
show(g)
```
